```python
import pandas as pd

import pandas as pd
from datetime import datetime

# Assuming df_metro, cutoff_3yr, cutoff_1yr, cutoff_3mo are defined

df_metro["PERIOD_BEGIN"] = pd.to_datetime(df_metro["PERIOD_BEGIN"])

def summarize_period(df, start, label):
    # 1. Filter the DataFrame first (as you did before)
    d = df[df["PERIOD_BEGIN"] >= start]

    # 2. Define all aggregations in a single dictionary
    agg_funcs = {
        "MEDIAN_SALE_PRICE": "mean",
        "MEDIAN_DAYS_ON_MARKET": "mean",
        "ACTIVE_LISTINGS": "mean",
        "PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS": "mean",
        "AVERAGE_SALE_TO_LIST_RATIO": "mean",
        "WEEKS_OF_SUPPLY": "mean",
        # Calculate TOTAL_SALES using 'sum' in the same groupby
        "ADJUSTED_AVERAGE_HOMES_SOLD": "sum",
    }

    # 3. Apply the combined aggregation
    summary = (
        d.groupby("REGION_NAME")
         .agg(agg_funcs) # Use the defined aggregation dictionary
         .reset_index()
    )

    # 4. Rename the summed column to TOTAL_SALES
    summary.rename(
        columns={"ADJUSTED_AVERAGE_HOMES_SOLD": "TOTAL_SALES"},
        inplace=True
    )
    
    # 5. Add the period label
    summary["Period"] = label
    
    return summary
df_5yr = summarize_period(df_metro, cutoff_5yr, "Past 5 Years")
df_3yr = summarize_period(df_metro, cutoff_3yr, "Past 3 Years")
df_1yr = summarize_period(df_metro, cutoff_1yr, "Past 1 Year")
df_3mo = summarize_period(df_metro, cutoff_3mo, "Past 3 Months")

combined = pd.concat([df_5yr,df_3yr, df_1yr, df_3mo])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

results = []
for label, group in combined.groupby("Period"):
    g = group.copy()
    metrics = [
        "MEDIAN_SALE_PRICE",
        "ACTIVE_LISTINGS",
        "MEDIAN_DAYS_ON_MARKET",
        "PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS",
        "TOTAL_SALES",
        "AVERAGE_SALE_TO_LIST_RATIO"
    ]
    g[metrics] = scaler.fit_transform(g[metrics])
    g["iBuyer_Score"] = (
        0.2 * (1 - g["MEDIAN_DAYS_ON_MARKET"]) +
        0.1 * g["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"]+
        0.2 * g["AVERAGE_SALE_TO_LIST_RATIO"]+
        0.3 * g["TOTAL_SALES"]
    )
    results.append(g)

metro_ranked = pd.concat(results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
top_metros = metro_ranked[metro_ranked["REGION_NAME"].isin([
    "Houston, TX metro area",
    "Atlanta, GA metro area",
    "Chicago, IL metro area",
    "Phoenix, AZ metro area",
    "Dallas, TX metro area",
    "Minneapolis, MN metro area"
])]

# --- Define Stage-2 derived metric computation ---
def compute_derived_metrics(df):
    d = df.copy()

    # --- Derived market indicators ---
    d["Turnover_Rate"] = d["TOTAL_SALES"] / d["ACTIVE_LISTINGS"]
    d["Market_Pressure"] = d["AVERAGE_SALE_TO_LIST_RATIO"] / d["WEEKS_OF_SUPPLY"]
    d["Demand_Strength"] = (
        d["AVERAGE_SALE_TO_LIST_RATIO"] / d["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"]
    )
    d["Liquidity_Score"] = 1 / d["MEDIAN_DAYS_ON_MARKET"]

    # --- Composite Health index (weighted) ---
    d["Composite_Health"] = (
        0.30 * d["Liquidity_Score"]
        + 0.25 * d["Turnover_Rate"]
        + 0.25 * d["Market_Pressure"]
        + 0.20 * d["Demand_Strength"]
    )

    # remove inf / outliers
    d.replace([np.inf, -np.inf], np.nan, inplace=True)
    for c in ["Turnover_Rate", "Market_Pressure", "Demand_Strength"]:
        d[c] = np.clip(d[c], None, d[c].quantile(0.99))

    return d

# --- Apply for each period ---
derived_results = []
for label, g in top_metros.groupby("Period"):
    d = compute_derived_metrics(g)
    d["Period"] = label
    derived_results.append(d)

derived_all = pd.concat(derived_results, ignore_index=True)

# --- Get top metros per period based on Composite_Health ---
for p in ["Past 5 Years", "Past 3 Years", "Past 1 Year", "Past 3 Months"]:
    top = (
        derived_all[derived_all["Period"] == p]
        .sort_values("Composite_Health", ascending=False)
        .head(5)
    )
    print(f"\n🏡 Top 5 Derived-Health Metros — {p}")
    print(
        top[
            [
                "REGION_NAME",
                "Composite_Health",
                "Liquidity_Score",
                "Turnover_Rate",
                "Market_Pressure",
                "Demand_Strength",
            ]
        ].to_markdown(index=False)
    )

# --- Visualization helper ---
def plot_bar(period_label):
    dfp = (
        derived_all[derived_all["Period"] == period_label]
        .sort_values("Composite_Health", ascending=False)
        .head(5)
    )
    plt.figure(figsize=(8, 5))
    plt.barh(dfp["REGION_NAME"], dfp["Composite_Health"], color="teal", alpha=0.8)
    plt.xlabel("Composite Health Score")
    plt.ylabel("Metro")
    plt.title(f"Composite Market Health — {period_label}")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

# Plot for each period
for p in ["Past 3 Years", "Past 1 Year", "Past 3 Months"]:
    plot_bar(p)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# <-- edit your shortlist here
TOP_METROS = [
    "Houston, TX metro area",
    "Atlanta, GA metro area",
    "Chicago, IL metro area",
    "Phoenix, AZ metro area",
    "Dallas, TX metro area",
    "Minneapolis, MN metro area",
]

df = df_metro.copy()
df["PERIOD_BEGIN"] = pd.to_datetime(df["PERIOD_BEGIN"])
df = df[df["REGION_NAME"].isin(TOP_METROS)].sort_values(["REGION_NAME","PERIOD_BEGIN"])
```

```python
# 8-week smoother (tweak if you prefer 4 or 12)
ROLL = 8

def smooth(s):
    return s.rolling(ROLL, min_periods=1).mean()

def pct_change_weeks(s, weeks=52):
    return (s / s.shift(weeks) - 1) * 100

def zscore(s):
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def cooling_heating_block(g):
    """
    Composite: Up when market is cooling (harder to sell), down when heating (easier).
    We invert signs so: positive = cooling, negative = heating.
    Components:
      + ΔDOM (up => cooling)
      + ΔWOS (up => cooling)
      + Δ%PriceDrops (up => cooling)
      - ΔSale-to-List (down => cooling)
    """
    dom = smooth(g["MEDIAN_DAYS_ON_MARKET"])
    wos = smooth(g["WEEKS_OF_SUPPLY"])
    pdrop = smooth(g["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"])
    stl = smooth(g["AVERAGE_SALE_TO_LIST_RATIO"])

    # 8-week momentum of each metric
    dom_mom  = dom.diff(ROLL)
    wos_mom  = wos.diff(ROLL)
    pdr_mom  = pdrop.diff(ROLL)
    stl_mom  = stl.diff(ROLL)

    comp = zscore(dom_mom) + zscore(wos_mom) + zscore(pdr_mom) - zscore(stl_mom)
    return comp.rename("COOLING_INDEX")

def add_summaries(g):
    g = g.copy()
    # smoothed levels
    g["Price_s"]   = smooth(g["MEDIAN_SALE_PRICE"])
    g["DOM_s"]     = smooth(g["MEDIAN_DAYS_ON_MARKET"])
    g["WOS_s"]     = smooth(g["WEEKS_OF_SUPPLY"])
    g["STL_s"]     = smooth(g["AVERAGE_SALE_TO_LIST_RATIO"])
    g["PDROP_s"]   = smooth(g["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"])

    # YoY (if native _YOY cols exist we'll still compute from levels for consistency)
    g["Price_YoY"] = pct_change_weeks(g["Price_s"], 52)
    g["DOM_YoY"]   = pct_change_weeks(g["DOM_s"], 52)
    g["WOS_YoY"]   = pct_change_weeks(g["WOS_s"], 52)
    g["STL_YoY"]   = pct_change_weeks(g["STL_s"], 52)
    g["PDROP_YoY"] = pct_change_weeks(g["PDROP_s"], 52)

    # Cooling/Heating composite
    g["COOLING_INDEX"] = cooling_heating_block(g)
    return g

def last_12m_slope(x):
    # slope over last ~52 weeks (nan-safe)
    x = x.dropna()
    if len(x) < 8: 
        return np.nan
    y = x.values[-52:] if len(x) >= 52 else x.values
    X = np.arange(len(y))
    m = np.polyfit(X, y, 1)[0]
    return m
```

```python
hist = (
    df.groupby("REGION_NAME", group_keys=True)
      .apply(add_summaries)
      .reset_index(drop=True)
)
```

```python
# Average market cycle length for top metros

import pandas as pd
import numpy as np

# Assuming you already have df_metro with 'REGION_NAME', 'PERIOD_BEGIN', etc.
df = df_metro.copy()
df = df[df["REGION_NAME"].isin(TOP_METROS)]
df["PERIOD_BEGIN"] = pd.to_datetime(df["PERIOD_BEGIN"])
df = df.sort_values(["REGION_NAME", "PERIOD_BEGIN"])

# Compute Cooling/Heating Index again (if needed)
def zscore(s):
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def compute_cooling_index(g):
    dom = g["MEDIAN_DAYS_ON_MARKET"].rolling(3, min_periods=1).mean()
    wos = g["WEEKS_OF_SUPPLY"].rolling(3, min_periods=1).mean()
    drops = g["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"].rolling(3, min_periods=1).mean()
    sale_ratio = g["AVERAGE_SALE_TO_LIST_RATIO"].rolling(3, min_periods=1).mean()

    # Weighted composite (higher = cooling)
    index = (0.4 * dom.rank(pct=True)
             + 0.3 * wos.rank(pct=True)
             + 0.2 * drops.rank(pct=True)
             - 0.1 * sale_ratio.rank(pct=True))
    return index.rename("COOLING_INDEX")

df["COOLING_INDEX"] = df.groupby("REGION_NAME", group_keys=False).apply(compute_cooling_index)





```

```python
def estimate_cycle_length(series, dates):
    """Estimate average market cycle in months, given value series & actual date column."""
    s = pd.Series(series.values, index=pd.to_datetime(dates, errors='coerce')).dropna()
    s = s.rolling(5, min_periods=1).mean()
    if len(s) < 12:
        return np.nan

    from scipy.signal import find_peaks
    peaks, _ = find_peaks(s, distance=3, prominence=0.05)
    troughs, _ = find_peaks(-s, distance=3, prominence=0.05)

    if len(peaks) < 2 and len(troughs) < 2:
        return np.nan

    def mean_months(idxs):
        if len(idxs) <= 1:
            return np.nan
        diffs_days = np.diff(s.index[idxs]).astype('timedelta64[D]').astype(float)
        return np.mean(diffs_days) / 30.44

    peaks_months = mean_months(peaks)
    troughs_months = mean_months(troughs)
    return np.nanmean([peaks_months, troughs_months])
```

```python
cycle_summary = (
    df.groupby("REGION_NAME")
      .apply(lambda g: estimate_cycle_length(
          g["COOLING_INDEX"], g["PERIOD_BEGIN"]))
      .reset_index(name="Avg_Market_Cycle_Length_Months")
      .sort_values("Avg_Market_Cycle_Length_Months", ascending=False)
)

print(cycle_summary.head(10).to_markdown(index=False))
```

```python
import pandas as pd
import numpy as np

# --- 1️⃣ Merge cycle metrics with Stage 1 summary ---
derived_all = derived_all[derived_all["Period"] == "Past 3 Years"]
merged = derived_all.merge(
    cycle_summary,
    on="REGION_NAME",
    how="left"
)

# --- 2️⃣ Clean up and derive stability-related features ---
merged["Avg_Market_Cycle_Length_Months"] = merged["Avg_Market_Cycle_Length_Months"].replace(0, np.nan)
merged["Cycle_Stability"] = 1 / merged["Avg_Market_Cycle_Length_Months"]
merged["Cycle_Stability"].fillna(merged["Cycle_Stability"].mean(), inplace=True)

# --- 3️⃣ Normalize relevant metrics for comparability ---
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

merged["Liquidity_norm"] = normalize(merged["Liquidity_Score"])
merged["Distress_norm"] = normalize(1 - merged["PERCENT_ACTIVE_LISTINGS_WITH_PRICE_DROPS"])  # reverse distress
merged["Volume_norm"] = normalize(merged["TOTAL_SALES"])
merged["Cycle_Stability_norm"] = normalize(merged["Cycle_Stability"])

# --- 4️⃣ Compute Stability-Adjusted iBuyer Opportunity Score ---
merged["Stage2_StabilityScore"] = (
    0.35 * merged["Liquidity_norm"] +          # faster sales
    0.25 * merged["Volume_norm"] +             # larger markets
    0.20 * merged["Cycle_Stability_norm"] +    # stable/predictable cycles
    0.20 * merged["Distress_norm"]             # fewer distressed listings
)

# --- 5️⃣ Rank metros ---
final_rank = merged.sort_values("Stage2_StabilityScore", ascending=False)

print("🏙️ Stability-Adjusted iBuyer Opportunity — Top 10 Metros")
print(final_rank[[
    "REGION_NAME", "Stage2_StabilityScore",
    "Liquidity_norm", "Volume_norm",
    "Distress_norm", "Cycle_Stability_norm",
    "Avg_Market_Cycle_Length_Months"
]].head(10).to_markdown(index=False))
```
